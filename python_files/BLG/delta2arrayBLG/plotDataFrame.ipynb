```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import sys, os
path_to_csv = './csvresults/'
```

```python
# %matplotlib notebook
NUMGS = 8
csv_names = [f'f{i}_complete_grid_results.csv' for i in range(NUMGS)]

dflist = []
for i, name in enumerate(csv_names):
    try:
        df = pd.read_csv(os.path.join(path_to_csv,name))
        dflist.append(df)
    except FileNotFoundError:
        print(f'csv file {name} not found')
    

```

```python
df1 = dflist[1]
omegavals1 = np.array(df1.columns[1:],dtype=np.float64)
rvals1 = np.array(df1.index)
r_idx1 = 1
nldos1 = (np.array(df1.iloc[r_idx1][1:]))
# plt.plot(omegavals1, nldos1)

# plt.loglog(omegavals, nldos)
plt.loglog(omegavals1, np.abs(nldos1))
```
